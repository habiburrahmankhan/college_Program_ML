### KNN Algo 

```python
import numpy as np
import matplotlib.pyplot as plt  
import pandas as pd 
```

```python
dfx=pd.read_csv("xdata.csv")
dfy = pd.read_csv("ydata.csv")
```

```python
X = dfx.values
Y = dfy.values
```

```python

print(X.shape)
print(Y.shape)
```

```python
X = X[:, 1:]
Y=Y[: , 1:].reshape((-1,))
print(X.shape)
print(Y.shape)
```

```python
plt.scatter(X[: , 0] ,X[: , 1] , c=Y)
plt.show()
```

```python
query_x = np.array([1,1])
plt.scatter(X[: , 0] , X[: , 1] ,c=Y )
plt.scatter(query_x[0] , query_x[1] , color="red")
plt.show()
```

```python
def dist(x1 , x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(X , Y , queryPoint , k=3):
    vals=[]
    m = X.shape[0]
    for i in range(m):
        d = dist(queryPoint , X[i])
        vals.append((d,Y[i]))
    vals = sorted(vals)
    vals = vals[:k]
    vals = np.array(vals)
    print(vals)
    new_vals = np.unique(vals[: , 1 ] , return_counts=True)
    index = new_vals[1].argmax()
    pred = new_vals[0][index]
    print(new_vals)
    return pred
```

```python
pred = knn(X , Y , query_x)
```

```python
print(pred)
```

```python
a = np.array([1 , 2 , 3 ])
b =  np.array([2 , 3 , 4 ])
print(dist(a , b))
```
